# 1 A simple plot

## 1.1 Letâ€™s look at some code.

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

plt.plot(x, y_1, label = 'Y values',
        color = 'red', linestyle = 'dashed')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha = .25)
plt.legend(loc = 'upper left')
```

### Things to note

```python
plt.plot(x, y_1, color = 'red', linestyle = 'dashed', linewidth = 2)
```

```python
plt.plot(x, y_1, c='red', ls='dashed', lw=2)
```

```python
plt.plot(x, y_1, y_2)
```

```python
plt.plot(x, y_1, label='Y values',
           color='red', linestyle='dashed')
```

## 1.2 Adding another plot

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed')
plt.plot(x, y_2, label='Y2 values', color='blue')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

## 1.3 Yet another plot but with error bars

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', ) #adding another line to the graph
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
             label='Y3 with errors', color='green') #adding error bars

plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

# 2 Better with NumPy

```python
import numpy as np
```

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
```

## 2.1 Adding mathematical functions

```python
x = np.array([1, 2, 3, 4, 5, 6])

x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label = 'x^2', color = 'red', linestyle = 'dashed',)
plt.plot(x, sin_x, label = 'sin(x)', color = 'blue')
plt.legend()
```

## We need another axis!

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label = 'x^2', color = 'red', linestyle = 'dashed')
plt.legend(loc = 'lower left')

plt.twinx() #creates a new y-axis

plt.plot(x, sin_x, label = 'sin(x)', color = 'blue',)
plt.legend(loc='lower right')
```

### Things to note

```python
x = np.linspace(0, 6, 100)
```

# 3 Saving to disc

```python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label = 'x^2', color = 'red', linestyle = 'dashed')
plt.legend(loc = 'lower left')
plt.twinx()
plt.plot(x, sin_x, label = 'sin(x)', color = 'blue')
plt.legend(loc = 'lower right')

plt.savefig('simple-plot.png', dpi = 150) #dpi is dots per inch
```

# 4 A real example: Global Warming

## 4.1 Plotting data from files

```python
data = np.loadtxt('Complete_TAVG_daily.txt', skiprows = 24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha = 0.5)
plt.ylim([-8, 8])
```

```python
data = np.loadtxt('Complete_TAVG_daily.txt', skiprows = 24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha = 0.5)
plt.ylim([-8, 8])

plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
```

## 4.2 Styles

```python
plt.style.use('fivethirtyeight')

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.ylim([-8, 8])

plt.xlabel('Date')
plt.ylabel('Temperature anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
```
